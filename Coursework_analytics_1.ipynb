### 1.0 Начальная подготовка данных

Импорт билиотек NumPy, Pandas, Matplotlib, Seaborn для дальнейшей работы

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime, date

warnings.filterwarnings('ignore')
```

Загрузка данных из CSV файлов

```python
courses = pd.read_csv('courses.csv', index_col=0)
students = pd.read_csv('students.csv')
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv', parse_dates=['start_date', 'finish_date'], dayfirst=False)

# Переименование столбцов в датафреймах для объединения
progresses.rename(columns={'id':'progress_id'}, inplace = True)   
courses.rename(columns={'id':'course_id'}, inplace = True) 
```

Объединение датафреймов

```python
main = progress_phases.merge(right=progresses) # по индентификатур прогресса

main = main.merge(right=courses, how='left') # по идентификатору курса

main = main.merge(right=course_contents, how='left') # объединяем по идентификатору курса
main.rename(columns={'title': 'course_title', 'field': 'course_field'}, inplace=True) # переименовываем столбцы
```

Проверка датафрейма

### 1.1 Расчет основных метрик

**1.1.1 Общее количество курсов**

```python
course_list = list(courses['title'])
print('\nКоличество курсов: {}\n'.format(len(course_list)))
print('Название курсов:\n', '\n'.join(course_list))
```

**1.1.2 Количество модулей на каждом курсе**

Подготавливаем отдельный датафрейм по всем курсам

```python
all_courses = course_contents.merge(courses, how='left')
all_courses.head(2)
```

```python
all_courses.groupby('title').agg(Number_of_module = ('module_number', 'max'))
```

**1.1.3 Количество уроков в каждом модуле на каждом курсе**

```python
# Группируем по курсам и по модулям, а потом считаем количество уроков в каждом модуле
lessons_count = all_courses.groupby(['title', 'module_title']).agg(Number_of_lessons=('lesson_title', 'count'))
lessons_count
```

**1.1.4 Медианное количество уроков в модуле на каждом курсе**

```python
# Группируем по курсам и считаем медианное значение количества уроков в модуле.
lessons_count.groupby('title').agg(Median_of_lessons=('Number_of_lessons', 'median'))
```

```python
# Медианное количество уроков во всех курсах
lessons_count.Number_of_lessons.median()
```

**1.1.5 Количество учеников на каждом курсе**

```python
# Для подсчета студентов на курсе группируем все по курсу, далее считаем количество уникальных Students_ID на каждом курсе
students_count = main.groupby(['course_field', 'course_title']).agg(Number_of_students = ('student_id', lambda x: len(x.unique())))
students_count
```

**1.1.6 Минимальный, максимальный, средний, медианный возраст студентов**

Подготавливаем датайрейм по студентам. Вычисляем возраст для тех стужентов у которых указана дата рождения.

```python
# Создаем функцию вычисления возраста по дате рождения и текущему дню
def get_age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

# Создаем новый датайрейм очищенный от пропущенных значений поля "Birthday"
students_age = students.dropna(subset=['id', 'birthday'])
students_age['age'] = students_age['birthday'].apply(get_age)

# Скидываем лишние столбцы
students_age.drop(['id_', 'city'], axis=1, inplace=True)
students_age.sort_values('age')
```

Видно, что датасет содержит некорректные значения даты рождения и возраста. Отфильтруем их.

```python
students_age = students_age[(students_age.age > 10) & (students_age.age < 90)]
students_age
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=students_age.age, bins=80)
plt.title('Распределение студентов по возрасту')
plt.xlabel('Возраст студентов, лет')
plt.ylabel('Количество студентов, чел.')
plt.show()
```

Вычисляем минимальный, максимальный и медианный возраст студентов

```python
print('Минимальный возраст студента:', students_age.age.min())
print('Максимальный возраст студента:', students_age.age.max())
print('Медианный возраст студентов:', np.median(students_age['age']))
```

Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

```python
students_courses = main[['course_title', 'student_id']]
students_courses = students_courses.merge(students_age, left_on='student_id', right_on='id', how='inner')
students_courses.groupby(['course_title', 'student_id']).agg('mean').reset_index()\
    .groupby('course_title').agg(min_age = ('age', min), max_age = ('age', max),
                                 mean_age = ('age', 'mean'), median_age = ('age', 'median'))
# Сгруппирова сначало по курсам потом по студентам мы получаем в колонке student_id только уникальные значения
# Поскольку возраст каждого студента одинаковый то при взятии среднего значения мы не теряем его.
```

### 2.1 Расчет прироста стедентов на каждом курсе

```python
# Создаем датайфрейм содержащий только домашние работы, сортируем по возрастанию модуля и урока
# Таким образом в первой строке оказывается минимальный номер модуля с бомашним заданием.
first_hw = all_courses[all_courses.is_homework == True].sort_values(['title', 'module_number', 'lesson_number'])

# Для каждого курса выбираем минимальный модуль с домашней работой.
first_hw = first_hw.loc[first_hw.groupby('title').module_number.idxmin()].reset_index()[['lesson_token', 'is_homework']]

# Переименовываем столбец для последующего слияния
first_hw.rename(columns={'is_homework':'first_homework'}, inplace = True)
# Таким образом, получили датафрейм содержащий токены первых домашних работ для каждого курса
```

```python
# Создаем новый датафрейм содержащий только данные по первой домашней работе на курсе 
increase = main.merge(first_hw, how='left' )
increase = increase[increase.first_homework == True]

# Создаем колонку Год - Месяц начала курса 
increase['Date'] = increase['start_date'].map(lambda x: datetime.strftime(x,'%Y-%m'))

# Группируем по названию курса и дате начала, после чего считает количество уникальных студентов
increase = increase.groupby(['course_title', 'Date']).agg(students_count = ('student_id', 'nunique'))
# Так как один и тотже студент на одном и томже курсе не может 2 раза начать делать домашнее задание то полученные данные 
# Как раз и отображают прирост студентов на курсе.

# Избавляемся от многоуровневой таблицы и заполняем пропущенные значения нулями
increase = increase.unstack(level=0, fill_value=0)
increase = increase['students_count']
increase
```

### 2.2 График прироста студентов на каждом курсе

```python
# Фильтруем только только домашние работы и только со статусом "выполнены" и сразу выбираем только интересующие нас столбцы
progress_home_work = main[(main.is_homework == True) & (main.status == 'done')]\
                                [['progress_id','course_title', 'module_number', 'start_date', 'finish_date']]

# Группируем по ID прогресса, оставляем название курса и выбираем минимальную дата начала и максимальную дату окончания
# Сбрасываем индексы и сотрируем по дате окончания модуля
progress_home_work = progress_home_work.groupby(['progress_id', 'module_number', 'course_title'])\
                     .agg(start_date = ('start_date', 'min'), finish_date= ('finish_date', 'max'))\
                     .reset_index()\
                     .sort_values('finish_date')

# Создаем колонку с периодом времени в течении которого выполнялась работа
progress_home_work['Month'] = progress_home_work.apply(lambda x: pd.period_range(x['start_date']\
                                                                               , x["finish_date"]\
                                                                               , freq="M")\
                                                       , axis=1)
# Формируем столбец содержащий список месяцев в формате ГГГГ-ММ в четении которых выполнялась работа
progress_home_work['Month'] = progress_home_work.Month.apply(lambda x: list(x.strftime('%Y-%m')))

# Сбрасываем ненужные теперь столбцы начала и конца работы
progress_home_work.drop(['start_date', 'finish_date'], axis=1, inplace=True)

# Разбиваем список месяцев в строки
progress_home_work = progress_home_work.explode('Month').sort_values('Month')

# Группируем по курсу и месяцу выполнения, считаем количество задача
progress_home_work = progress_home_work.groupby(['course_title', 'Month']).agg(total_tasks = ('progress_id', 'count'))

# Формируем новый датафрейм в виде сетки из месяцов и курсов, пропущенные значения заполняем нулями
progress_home_work = progress_home_work.unstack(level=0, fill_value=0)
total_tasks = progress_home_work['total_tasks']
total_tasks
```

### 3.1 Расчет минимального, максимального и среднего времени проходжения каждого модуля для каждого курса

```python
# формируем новый датасет time_for_module Группируем по курсу модулю и студенту
# сбрасываем индексы и удаляем пропущенные значения
time_for_module = main.dropna()\
                        .groupby(['course_title', 'module_title', 'module_number', 'progress_id'])\
                        .agg(start = ('start_date', 'min')\
                            ,stop = ('finish_date', 'max'))\
                        .reset_index()\

# Вычисляем время выполнения заданий для каждого студента округляем значения до целых дней и переводим TimeDelta в INT
time_for_module['run_time'] = (time_for_module.stop - time_for_module.start).round('D').apply(lambda x: x.days)

# Группируем по курсу и модулю и считаем необходимые метрики
time_for_module = time_for_module.groupby(['course_title', 'module_title', 'module_number'])\
                                 .agg(min_time = ('run_time', 'min')\
                                     ,max_time = ('run_time', 'max')\
                                     ,mean_time = ('run_time', 'mean')\
                                     ,median_time = ('run_time', 'median'))\
                                 .sort_values(['course_title', 'module_number'])
                                
time_for_module
```

### 4.1 Расчет конверсии перехода стендов из одного модуля в другой на каждом курсе

```python
# Из основного датафрейма выбираем только записи с выполненными домашними работами
# группируем по курсам студентам и модулям выбираем даты старта и окончания домашних работ с минимальными значениями
st_conv_rate = main[(main.is_homework == True)]
st_conv_rate = st_conv_rate.groupby(['course_title', 'student_id', 'module_title', 'module_number', 'status'])\
                           .agg(start_date = ('start_date', 'min'), finish_date = ('finish_date', 'min'))\
                           .reset_index()

# Считаем количество студентов всех статусов выполнения работ для каждого модуля в каждом курсе
# сортируем по возрастанияю номера модуля, заполняем пропущенные значения нулями
# Для удобства разворачиваем индекс 3го уровня в столбцы
st_conv_rate = st_conv_rate.groupby(['course_title', 'module_title', 'module_number', 'status'])\
                           .agg(total_students = ('student_id', 'count'))\
                           .unstack(level='status', fill_value=0)\
                           .sort_values(['course_title', 'module_number'])

# Убираем 1 уровень
st_conv_rate = st_conv_rate['total_students']

# Создаем колонку с количеством студентов которые начали выполнение задания
st_conv_rate['all'] = st_conv_rate.sum(axis=1)

# Смещаем столбец done на 1 стороку вниз, удаляем индекс номера модуля и убираем все первые уроки из датафрейма
st_conv_rate['last_done'] = st_conv_rate.done.shift(1)
st_conv_rate = st_conv_rate.reset_index('module_number')
aaa = st_conv_rate

# Формируне название первых модулей каждого курса
first_module = []
for course in course_list:
     first_module.append(st_conv_rate.loc[course].index[0])

# Формируем список уникальных имен первых модулей из названия курса и названия модуля и удаляем эти строки из датайрейма
for i_index in zip(course_list, first_module):
     st_conv_rate.drop(index=i_index, inplace=True)
first_module


# Расчитываем конверсию
st_conv_rate['conv'] = st_conv_rate['all'] / st_conv_rate['last_done']
st_conv_rate.head(50)
```

### 5.1 Оценка по количеству заброшенных заданий

Для поиска проблемных студентов с низкой мотивацией или низкой первоначальной подготовкой целесообразным видеться посчитать студентов у которых количество начатых и не законченных заданий превышает или равно 3 шт.

```python
# Фильтруем из основного датасета все задния домашних работсо статусом fail and start
fail_st = main[(main.is_homework == True)]
fail_st = fail_st[(fail_st.status == 'fail') | (fail_st.status == 'start')][['course_title', 'student_id', 'status']]


# Группируем по курсу и студенту и считаем общее количество начатых и невыполненных заданий, сортируем по убыванию
fail_st = fail_st.groupby(['course_title', 'student_id'])\
                 .agg(total_fail = ('status', 'count'))\
                 .sort_values(['course_title', 'total_fail'], ascending=False)\

# Выводим всех студентов у которых количество "висящих" заданий 3 и более
fail_st = fail_st[fail_st.total_fail >= 3]
fail_st
```

```python
problem_course = fail_st.reset_index().groupby(['course_title'])\
                                      .agg(problem_students_count = ('student_id', 'count'))\
                                      .sort_values('problem_students_count', ascending=False)

problem_course
```
